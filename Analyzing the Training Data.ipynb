### Analyzing the Training Data
We will use the moving average method to smooth the curves in order to make the data more readable.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the training data
data_train = pd.read_csv('X_train_Wwou3IE.csv')

# Drop "predicted_spot_price" column
data_train = data_train.drop(columns=["predicted_spot_price"])
```

```python
data_train['DELIVERY_START'] = pd.to_datetime(data_train['DELIVERY_START'], utc=True)  # Parse with timezone info
data_train = data_train.set_index('DELIVERY_START')
```

```python
# Check the proportion of missing values in each column
round(data_train.isna().sum()/data_train.shape[0]*100,2)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['load_forecast'].rolling(window=50, center=True).mean(),label='load_forecast')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['coal_power_available'].rolling(window=50, center=True).mean(),label='coal_power_available')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['gas_power_available'].rolling(window=50, center=True).mean(),label='gas_power_available')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['nucelear_power_available'].rolling(window=50, center=True).mean(),label='nucelear_power_available')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['wind_power_forecasts_average'].rolling(window=50, center=True).mean(),label='wind_power_forecasts_average')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['wind_power_forecasts_std'].rolling(window=50, center=True).mean(),label='wind_power_forecasts_average')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_train['solar_power_forecasts_std'].rolling(window=50, center=True).mean(),label='solar_power_forecasts_average')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```
